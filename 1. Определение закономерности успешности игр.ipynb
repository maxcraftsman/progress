```python
from collections import Counter

#Вспомогательная функция для get_all_modes(). Возвращает наиболее часто встречающееся значение из списка.
def most_frequent(List): 
    occurence_count = Counter(List) 
    return occurence_count.most_common(1)[0][0] 

#Вспомогательная функция для mode_group(). Получает все значения моды из кортежа.
def get_all_modes(a):
    c = Counter(a)  
    mode_count = max(c.values())
    mode = {key for key, count in c.items() if count == mode_count}
    cleanedList = [x for x in mode if str(x) != 'nan']
    if cleanedList == []:
        return 0
    else:
        return most_frequent(cleanedList)
    
#Вспомогательная функция для mode_nan(). Формирует запрос и обрабатывает результат.
def mode_group(df, val, col, group):
    temp_group = df.query('{} == "{}"'.format(group,val[group]))
    result = get_all_modes(temp_group[col])
    try:
        if result == 0:
            text_result = "Нет"
            return None
        else:
            text_result = result
            return result
    finally:
        print('{:>6} - установлено по моде для "{}"'.format(text_result, val['name']))

#Функция mode_nan() берёт колонку с пропущенными значениями и заполняет их по моде другой колонки.
#Принимает следующие параметры: датафрейм, колонка с пропущенными значениями, колонка-мода.
def mode_nan(df, target_col, mode_col):
    printmd('**Замена значений NaN в колонке "{}", по моде "{}":**'.format(target_col,mode_col))
    full_query = '{} != {}'.format(target_col,target_col)
    
    copy = df.query(full_query).apply(lambda x: mode_group(df,x,target_col,mode_col), axis=1)
    df_step_1[target_col].update(copy)

mode_nan(df_step_1,'year_of_release','name')
```

### 3.4 Сравнение продаж игр на разных платформах:

```python
#Изучил общее распределение игр по платформам. 
platforms = df_period.groupby('platform')['total_sales'].count().sort_values(ascending=False).to_frame()
platforms.columns = ['count']
platforms['percent'] = platforms['count'].apply(lambda x: round(x*100/platforms['count'].sum(),1))

#Записал в переменную plaforms суммы общих продаж по платформам.
sales_by_platform = df_period.groupby('platform')['total_sales'].sum().sort_values(ascending=False).to_frame()
platforms['total_sales'] = sales_by_platform['total_sales']
platforms = platforms[['count','total_sales','percent']]

printmd('**Cуммы общих продаж по платформам:**')
platforms.T
```

```python
#Функция game_describe() описывает продажи, на вход принимает: датафрейм, колонку и значения колонки, 
#по которым требуется разбить описания.
def game_describe(df,col,cat):
    temp_query = df.query('{} == @cat'.format(col))
    
    #Записал describe() для total_sales в переменную temp_desc
    temp_desc = df.query('{} == @cat'.format(col))['total_sales'].describe().to_frame() 
    temp_1q = temp_desc['total_sales']['25%'] # Первый квартиль
    temp_3q = temp_desc['total_sales']['75%'] # Третий квартиль
    temp_max = temp_desc['total_sales']['max'] # Максимальное значение
    highPlot = temp_3q + (temp_3q - temp_1q)*1.5 # Верхнее(правое) значение усов
    bestseller = df.query('{} == @cat and total_sales >= @highPlot'.format(col))['total_sales'].count() # Количество бестселлеров

    printmd('**{}**'.format(cat))
    print("{:<24} {:.2f}".format('Количество игр:',temp_query['total_sales'].count()))
    print("{:<24} {:.2f}".format('Общие продажи:',temp_query['total_sales'].sum()))
    print("{:<24} {:.2f} млн. копий".format('Макс. продажа:',temp_max))
    print("{:<24} {:.2f} млн. копий".format('Верхний ус продаж:',highPlot))
    print("{:<24} {:.2f} млн. копий".format('Среднее значение:',temp_query['total_sales'].mean()))
    print("{:<24} {:.0f}".format('Дисперсия:',temp_query['total_sales'].var(ddof=0)))
    print("{:<24} {:.2f}".format('Стандартное отклонение:',temp_query['total_sales'].std(ddof=0)))
    print("{:<24} {:.2f}%".format('Шанс релиза бестселлера:',(bestseller*100) / temp_query['total_sales'].count()))
    
#Построил «ящики с усами» по глобальным продажам игр в разбивке по платформам.
def plotbox_platform(df):
    plt.subplots(figsize=(16, 16))
    plt.subplots_adjust(hspace = 0.5)
    plt_i = 421

    for platform in actual_platforms:
        plt.subplot(plt_i).title.set_text(platform)
        temp_query = df.query('platform == @platform')

        temp_query.boxplot(column='total_sales',fontsize=11)
        
        game_describe(df,'platform',platform)
        plt_i += 1


    plt.show()
    
plotbox_platform(df_period)
```

```python
reg_list = ['na_sales','eu_sales','jp_sales','other_sales']
region_dict = {'na_sales':['Доли продаж в Северной Америке'],'eu_sales':['Доли продаж в Европе'],'jp_sales':['Доли продаж в Японии'],'other_sales':['Доли продаж в других странах']}
region_top_sales = df_period.groupby('platform')[reg_list].agg('sum').reset_index()

def circle_hist(df,col,cat,tdict,legend):
    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(aspect="equal"), dpi= 90)
    
    merge = pd.concat([df[col],df[cat]],axis=1)
    merged = merge.sort_values(by=col,ascending=False)
    total_other = merged[5:].sum()[0]
    temp_df = pd.DataFrame({col:[total_other],cat:['other']})
    if total_other > 0:
        merged = pd.concat([merged[:5],temp_df])
    
    data = merged[col]
    categories = merged[cat]

    explode = []
    i = 0
    while i != merged[cat].count():
        explode.append(0)
        i += 1
 
    def func(pct, allvals):
        absolute = int(pct/100.*np.sum(allvals))
        return "{:.1f}%".format(pct)

    wedges, texts, autotexts = ax.pie(data, 
                                      autopct=lambda pct: func(pct, data),
                                      textprops=dict(color="w"), 
                                      colors=plt.cm.Dark2.colors,
                                     startangle=140,
                                     explode=explode)

    # Decoration
    ax.legend(wedges, categories, title=legend, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    plt.setp(autotexts, size=10, weight=700)
    ax.set_title(tdict[col][0])
    plt.show()
    merged = merged.rename({col: 'млн. копий'}, axis=1) 
    display(merged.set_index([cat], append=True))
    print()
    print()

for region in reg_list:
    circle_hist(region_top_sales,region,'platform',region_dict,'Платформы')
    
region_top_sales.set_index('platform', inplace=True)
```

### Сравниваем генеральные совокупности пользовательских рейтингов  для платформ Xbox One и PC.

<b>Нулевая гипотеза:</b> средние пользовательские рейтинги платформ Xbox One и PC одинаковые.<br>
<b>Альтернативную гипотезу:</b> средние пользовательские рейтинги платформ Xbox One и PC различаются.

```python
from scipy import stats as st

df_period_pc = df_period.query('platform == "PC" and user_score == user_score')
df_period_xbox = df_period.query('platform == "XOne" and user_score == user_score')

#fix_3
print('Среднее кол-во продаж игры для PC: {:>10.2f}'.format(df_period_pc['total_sales'].mean()))
print('Среднее кол-во продаж игры для Xbox One: {:.2f}'.format(df_period_xbox['total_sales'].mean()))
print()

print('Количество наблюдений ген. совокупности PC: {:>9}'.format(len(df_period_pc)))
print('Количество наблюдений ген. совокупности Xbox One: {}'.format(len(df_period_xbox)))
print()

#fix_3
print('Дисперсия выборки, PC: {:>10.2f}'.format(df_period_pc['total_sales'].var(ddof=0)))
print('Дисперсия выборки, Xbox One: {:.2f}'.format(df_period_xbox['total_sales'].var(ddof=0)))
print()

#Пороговое значение alpha установил 5%
alpha = .05
#fix_3 Дисперсии не равны, разница больше 5%, поэтому установил equal_var равным False.
results = st.ttest_ind(df_period_pc['user_score'], df_period_xbox['user_score'], equal_var = False)
print('p-значение:', results.pvalue) 

printmd('**Вывод:**')
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу. Средние пользовательские рейтинги платформ различаются.")
else:
    print("Не получилось отвергнуть нулевую гипотезу. Средние пользовательские рейтинги платформ Xbox One и PC равны.")
```

### Сравниваем генеральные совокупности пользовательских рейтингов для жанров Action и Sports.

<b>Нулевая гипотеза:</b> Средние пользовательские рейтинги жанров Action и Sports разные.<br>
<b>Альтернативную гипотезу:</b> Средние пользовательские рейтинги жанров Action и Sports одинаковые.

```python
df_period_action = df_period.query('genre == "Action" and user_score == user_score')
df_period_sports = df_period.query('genre == "Sports" and user_score == user_score')

#fix_3
print('Среднее кол-во продаж для Action игры: {:.2f}'.format(df_period_action['total_sales'].mean()))
print('Среднее кол-во продаж для Sports игры: {:.2f}'.format(df_period_sports['total_sales'].mean()))
print()

print('Количество наблюдений ген. совокупности Action: {}'.format(len(df_period_action)))
print('Количество наблюдений ген. совокупности Sports: {}'.format(len(df_period_sports)))
print()

#fix_3
print('Дисперсия выборки, Action: {:.2f}'.format(df_period_action['total_sales'].var(ddof=0)))
print('Дисперсия выборки, Sports: {:.2f}'.format(df_period_sports['total_sales'].var(ddof=0)))
print()

#Пороговое значение alpha установил 5%
alpha = .05
#fix_3 Дисперсии не равны, разница больше 5%, поэтому установил equal_var равным False.
results = st.ttest_ind(df_period_action['user_score'], df_period_sports['user_score'], equal_var = False)
print('p-значение:', results.pvalue) 

printmd('**Вывод:**')
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу. Средние пользовательские рейтинги жанров Action и Sports различаются.")
else:
    print("Не получилось отвергнуть нулевую гипотезу. Средние пользовательские рейтинги жанров Action и Sports одинаковые.")
```
